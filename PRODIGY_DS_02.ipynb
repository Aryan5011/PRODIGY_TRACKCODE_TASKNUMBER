```python
data.hist(figsize=(10,9),color='skyblue')
plt.show()
```

```python
a1 = data['Sex'].value_counts()
a1
```

```python
plt.figure(figsize=(5,5))
plt.pie(a1.values,labels=a1.index,autopct='%.2f%%',colors=['blue','yellow'])
plt.title('Percentage Distribution of Sex Column')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.pie(data['Survived'].value_counts(),labels=data['Survived'].value_counts(),autopct='%.2f%%',colors=['orange','lightgreen'])
plt.title('Percentage Distribution of Survival Column')
plt.legend()
plt.show()
```

```python
classs = data.groupby('Pclass')['Survived'].mean()
print(classs)
```

```python
plt.figure(figsize=(5,6))
sns.countplot(x=data['Survived'], palette='pink')
data['Survived'].value_counts()
```

```python
sns.countplot(x='Sex',hue='Survived',data=data, palette='rocket')
plt.title('Distribution of Sex Column by Survival')
plt.show()
```

```python
classs = data.groupby('Pclass')['Survived'].mean()
print(classs)
```

```python
plt.bar(classs.index, classs.values)
plt.xlabel('Pclass',fontsize = 15)
plt.ylabel('Survival Rate',fontsize = 15)
plt.title('Survival Rate by Pclass',fontsize = 25)
plt.show()
```

```python
counts = data.groupby(['Pclass', 'Survived']).size().unstack()
counts.plot(kind="bar")
plt.show()
```

```python
sns.countplot(x='Pclass',hue='Sex',data=data, palette='viridis')
plt.title('Distribution of Pclass Column by Gender')
plt.show()
```

```python
sns.histplot(x=data['Age'],kde=True,color='r')
plt.title('Age Distribution')
plt.show()
```
